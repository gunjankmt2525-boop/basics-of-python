check if a number is positive or negetive

```python
num = 5

if num > 0:
    print("Positive number")
else:
    print("Negative number")
```

check positive,negetive or zero

```python
num = 0
if num > 0:
    print("Positive number")
elif num < 0:
    print("Negative number")
else:
    print("Zero")
```

grade system using elif

```python
marks = 85
if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

check leap year

```python
year = 2024
if year % 4 == 0:
    print("Leap year")
else:
    print("Not a leap year")
```

check if number is even

```python
num = 10
if num % 2 == 0:
    print("Number is even")
```
