Print number from 1 to 10

```python
for i in range(1, 11):
    print(i)
```

Print even number from 1 to 20





```python
for i in range(1, 21):
    if i % 2 == 0:
        print(i)
```

Print table of a number

```python
num = 5
i = 1

while i <= 10:
    print(num, "x", i, "=", num * i)
    i += 1
```

Print sum of first 10 numbers

```python
i = 1
sum = 0

while i <= 10:
    sum = sum + i
    i += 1

print("Sum =", sum)
```
